# **Business Understanding**

### About the Data

This dataset contains information about smartphone usage and system performance that influence battery drain. It includes features such as screen brightness, CPU usage, RAM usage, network connectivity, app activity, and device temperature.

The dataset can be used for exploratory data analysis, battery consumption studies, and machine learning models to understand and predict how different factors impact smartphone battery life.

[Source](https://www.kaggle.com/datasets/shayanzk/smartphone-battery-drain-and-usage-behavior?select=smartphone_battery_drain_dataset.csv)
