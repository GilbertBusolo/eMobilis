# **Business Understanding**

**About Dataset**
This dataset captures general online learning behavior across different age groups and roles, focusing on how learners use various online platforms, devices, and content formats.

It includes information such as preferred learning platforms, weekly learning time, learning purpose, content format preference, completion rates, and overall satisfaction scores. The dataset is designed to be easy to understand, privacy-safe, and suitable for both beginners and experienced data practitioners.


<hr> 

### **Source(Provenance)**

[Source](https://www.kaggle.com/datasets/prince7489/online-learning-platform-usage-dataset?resource=download)


**Disclaimer**
This dataset is synthetically generated to represent common patterns in online learning behavior. The structure and variable selection are inspired by:

- General trends observed on popular online learning platforms
- Publicly discussed online education usage reports
- Typical learner behavior scenarios across students and working professionals
- No real user data was collected, and no personally identifiable information is included.

**Collection Methodology**
The data was generated programmatically using controlled randomization to simulate realistic learning behavior. Methodology details: Each row represents a hypothetical learner profile Age groups, roles, platforms, and devices were sampled from predefined categories Weekly learning hours, completion rates, and satisfaction scores were generated within realistic ranges Logical constraints were maintained (e.g., completion rates between 10–100%, satisfaction on a 1–10 scale)

#### **Basic Visualization**
- pie chart 

```python
# plots pie chart
def plotPieChart(labels, data):
    fig, ax = plt.subplots()
    ax.pie(data, labels=labels)

# returns the labels and data for plotting
def getLabelData(df, targetColumn):
    labels =df[targetColumn].unique()
    data = list(df[targetColumn].value_counts())
    return labels, data
```

```python
label, data = getLabelData(learning_df, 'age_group')
plotPieChart(label, data )
```

```python
label, data = getLabelData(learning_df, 'role')
plotPieChart(label, data )
```

```python
label, data = getLabelData(learning_df, 'primary_platform')
plotPieChart(label, data )
```
